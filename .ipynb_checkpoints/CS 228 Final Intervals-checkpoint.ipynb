```python
import numpy as np

class interval:
    def __init__(self, dimension, lower, higher):
        self.dimension = dimension
        self.lower = lower
        self.higher = higher
    
    def classify(self, x):        
        for i, val in enumerate(x):
            if val > self.higher[i] or val < self.lower[i]:
                return False
            
        return True
    
    def get_dimension(self):
        return self.dimension
    
    def get_upper_bounds(self):
        return self.higher
    
    def get_lower_bounds(self):
        return self.lower
```

```python
import numpy as np

class noisy_interval:
    def __init__(self, dimension, lower, higher, eta):
        self.dimension = dimension
        self.lower = lower
        self.higher = higher
        self.eta = eta
        self.total_count = 0
        self.error_count = 0
    
    def classify(self, x):    
        inside = True
        self.total_count += 1
        for i, val in enumerate(x):
            if val > self.higher[i] or val < self.lower[i]:
                inside = False
        
        rand = np.random.uniform(0, 1)   
        if self.total_count % 1000 == 0:
            print(self.total_count)
        
        if rand < self.eta:
            self.error_count += 1
            return not inside
        
        return inside
    
        
    def get_dimension(self):
        return self.dimension
    

    def get_upper_bounds(self):
        return self.higher
    
    def get_lower_bounds(self):
        return self.lower
    
    def get_counts(self):
        return self.total_count, self.error_count
```

```python
def sprt(phi, a, r, curr_S):
    return curr_S + phi * np.log(a/r) + (1-phi) * np.log((1-a) / (1-r))
```

```python
def make_good_interval(noisy_interval):
    dimension = noisy_interval.get_dimension()
    lower = noisy_interval.get_lower_bounds()
    upper = noisy_interval.get_upper_bounds()
    
    return interval(dimension, lower, upper)
```

```python
def get_sample_point(dimension):
    return np.array(np.random.uniform(-100, 100, dimension))
```

```python
def get_finite_interval(dimension):
    bounds = np.round(np.random.uniform(-100, 100, (dimension, 2)), decimals=2)
    lower = np.min(bounds, axis=1)
    upper = np.max(bounds, axis=1)
    
    return interval(dimension, lower, upper)
    
```

```python
def get_finite_noisy_interval(dimension, eta):
    bounds = np.round(np.random.uniform(-100, 100, (dimension, 2)), decimals=2)
    lower = np.min(bounds, axis=1)
    upper = np.max(bounds, axis=1)
    
    return noisy_interval(dimension, lower, upper, eta)
    
```

```python
def get_noisy_bound(epsilon, delta, eta, dimension):
    return (2 / epsilon ** 2) / (1 - 2*eta) ** 2 * np.log(2 * (10 ** 4 * (10 ** 4 - 1) / 2) ** dimension /delta)
```

```python
def find_noisy_consistent_interval(examples, max_errors, test_cap, classifications, dimension):
    found = False
    # print(eta)
    length = len(examples)
    checks = 0
    while not found and checks <= test_cap:
        C = get_finite_interval(dimension)
        found = True
        errors = 0
        for i, example in enumerate(examples):
            if classifications[i] != C.classify(example):
                errors += 1
                if errors > max_errors:
                    # print(errors)
                    found = False
                    break
        checks += 1
    if found:
        return C, errors
    
    return None, -1
```

```python
def S(true_concept, epsilon, delta, eta, K = 2):
    greater = np.random.uniform(0, 1) > 0.5
    dimension = true_concept.get_dimension()
    C = get_finite_interval(dimension)
    hypotheses = np.array([C])
    deltas = np.array([6 * (delta) / np.pi ** 2])
    S_values = np.zeros(1)
    num_errors = np.zeros(1)
    min_errors = 0
    n = 1
    run = 0
    found = False
    concept = None
    examples = []
    classifications = []
    found_true = False
    found_false = False
    
    a = (epsilon - 2 * epsilon * eta) /K + eta
    r = epsilon + eta - 2 * epsilon * eta
    while not found:
        #print(len(examples))
        #print(n)
        run += 1
        example = get_sample_point(dimension)
        examples.append(example)
        curr_class = true_concept.classify(example)
        if curr_class:
            found_true = True
        else:
            found_false = True
        classifications.append(curr_class)
        for i in range(n):
            hyp = hypotheses[i]
            incorrect = 1
            if hyp.classify(example) == curr_class:
                incorrect = 0
                num_errors[i] += 1
            #print(sprt(incorrect, epsilon/K, epsilon, S_values[i]))
            S_values[i] = sprt(incorrect, a, r, S_values[i])
            #print(S_values[i])
        for i in range(n):
            if S_values[i] >= np.log(1/deltas[i]):
                found = True
                concept = hypotheses[i]
       # print(run)
        min_errors = np.min(num_errors)
        
        if hypotheses[-1].classify(example) != curr_class and (found_true and found_false):
            cap = 1000
            if eta != 0:
                cap = 2/eta
            new_hyp, error_num = find_noisy_consistent_interval(examples, min_errors, cap, classifications, dimension)
            if error_num != -1:
                hypotheses = np.append(hypotheses, new_hyp)
                n += 1
                # print(n)
                deltas = np.append(deltas, 6 * (delta) / (np.pi ** 2 * n ** 2))
                S_values = np.append(S_values, 0)
                num_errors = np.append(num_errors, error_num)
        
    return concept, examples
```

```python
import csv
import matplotlib.pyplot as plt

trials = 25
dimension = 1

f = open('1d-gamma-versus-sample-INTERVAL.csv', 'w')
writer = csv.writer(f)
writer.writerow(['eta', 'theoretical bound', 'practical sample'])

run = 0
bounds = []
records = []
means = []
for (i, curr_eta) in enumerate(etas):
    if curr_eta >= 0.3:
        trials = 10
    print(curr_eta)
    eta_record = []
    bounds.append(get_noisy_bound(epsilon, delta, curr_eta, dimension))
    for trial in range(trials):
        print(trial)
        C = get_finite_noisy_interval(dimension, curr_eta)
        
        found, examples = S(C, epsilon, delta, curr_eta)
        eta_record.append(len(examples))
    records.append(eta_record)
    means.append(np.mean(eta_record))
    row = [curr_eta, get_noisy_bound(epsilon, delta, curr_eta, dimension), np.mean(eta_record)]
    writer.writerow(row)
    plt.plot(etas[:i+1], bounds, label="Theoretical bound", color='blue')
    plt.plot(etas[:i+1], means, label="Practical Average", color='red')
    plt.xlabel("Eta Noise Rate")
    plt.ylabel("Sample Complexity")
    plt.title("Theoretical and Practical Sample Complexity of Interval\n Under Classification Noise")
    plt.legend()
    run += 1
    plt.savefig("Interval-Whole-With-Large-Eta" + str(run) + '.png')
    plt.show()
    plt.plot(etas[:i+1], means, label="Practical Average", color='red')
    plt.xlabel("Eta Noise Rate")
    plt.ylabel("Sample Complexity")
    plt.title("Practical Sample Complexity of Interval \n Under Classification Noise")
    plt.legend()
    plt.savefig("Interval-Practical-With-Large-Eta" + str(run) + '.png')
    plt.show()
        
```
