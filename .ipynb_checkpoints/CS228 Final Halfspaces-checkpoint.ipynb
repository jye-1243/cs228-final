```python
import numpy as np
import matplotlib.pyplot as plt

class halfspace:
    def __init__(self, dimension, params, greater, b):
        self.params = params
        self.dimension = dimension
        self.greater = greater
        self.b = b
    
    def classify(self, x):
        return (np.dot(self.params, x) > self.b) == self.greater
    
    def vc_dim(self):
        return self.dimension + 1
    
    def get_params(self):
        return self.params, self.greater, self.b
    
```

```python
class noisy_halfspace:
    def __init__(self, dimension, params, greater, b, eta=0):
        self.params = params
        self.dimension = dimension
        self.greater = greater
        self.b = b
        self.eta = eta
        self.total_count = 0
        self.error_count = 0
    
    def classify(self, x):
        self.total_count += 1
        rand = np.random.uniform(0, 1)   
        if self.total_count % 1000 == 0:
            print(self.total_count)
        if rand < self.eta:
            self.error_count += 1
            return (np.dot(self.params, x) > self.b) != self.greater
        
        return (np.dot(self.params, x) > self.b) == self.greater
    
    def vc_dim(self):
        return self.dimension + 1
    
    def get_params(self):
        return self.params, self.greater, self.b, self.eta
    
    def get_points(self):
        return self.points
    
    def get_counts(self):
        return self.total_count, self.error_count
```

```python
def get_theoretical_sample(epsilon, delta, d):
    a = 8 * d / epsilon * np.log(13/epsilon) / np.log(2)
    b = 4/epsilon * np.log(2/delta) / np.log(2)
    
    return max(a, b)
```

```python
def get_better_theoretical_sample(epsilon, delta, d):
    return 1 / (epsilon * (1- np.sqrt(epsilon))) * (2 * d * np.log(6/epsilon) + np.log(2/delta))
```

```python
def sprt(phi, a, r, curr_S):
    return curr_S + phi * np.log(a/r) + (1-phi) * np.log((1-a) / (1-r))
```

```python
def find_consistent_halfspace(examples, true_concept, dimension):
    found = False
    while not found:
        greater = np.random.uniform(0, 1) > 0.5
        C = get_finite_halfspace(dimension)
        found = True
        for example in examples:
            if true_concept.classify(example) != C.classify(example):
                found = False
                break 
    
    return C
```

```python
def find_noisy_consistent_halfspace(examples, max_errors, test_cap, classifications, dimension):
    found = False
    # print(eta)
    length = len(examples)
    checks = 0
    while not found and checks <= test_cap:
        greater = np.random.uniform(0, 1) > 0.5
        C = get_finite_halfspace(dimension)
        found = True
        errors = 0
        for i, example in enumerate(examples):
            if classifications[i] != C.classify(example):
                errors += 1
                if errors > max_errors:
                    # print(errors)
                    found = False
                    break
        checks += 1
    if found:
        return C, errors
    
    return None, -1
```

def find_noisy_consistent_halfspace(examples, classifications, dimension = 1):

    lr = LogisticRegression()
    lr.fit(examples, classifications)
    intercept = lr.intercept_[0] / lr.coef_[0]
    found_true = halfspace(1, np.array([1/intercept]), True, 1)
    found_false = halfspace(1, np.array([1/intercept]), False, 1)
    
    true_sum = 0
    false_sum = 0
    
    for i, example in enumerate(examples):
        if found_true.classify(example) == classifications[i]:
            false_sum += 1
        if found_false.classify(example) == classifications[i]:
            true_sum += 1
    
    if true_sum > false_sum:
        return found_true, len(examples) - true_sum
    
    return found_false, len(examples) - false_sum


```python
def S(true_concept, epsilon, delta, eta, K = 2):
    greater = np.random.uniform(0, 1) > 0.5
    dimension = true_concept.vc_dim() - 1
    C = get_finite_halfspace(dimension)
    hypotheses = np.array([C])
    deltas = np.array([6 * (delta) / np.pi ** 2])
    S_values = np.zeros(1)
    num_errors = np.zeros(1)
    min_errors = 0
    n = 1
    run = 0
    found = False
    concept = None
    examples = []
    classifications = []
    found_true = False
    found_false = False
    
    a = (epsilon - 2 * epsilon * eta) /K + eta
    r = epsilon + eta - 2 * epsilon * eta
    while not found:
        #print(len(examples))
        #print(n)
        run += 1
        example = get_sample_point(dimension)
        examples.append(example)
        curr_class = true_concept.classify(example)
        if curr_class:
            found_true = True
        else:
            found_false = True
        classifications.append(curr_class)
        for i in range(n):
            hyp = hypotheses[i]
            incorrect = 1
            if hyp.classify(example) == curr_class:
                incorrect = 0
                num_errors[i] += 1
            #print(sprt(incorrect, epsilon/K, epsilon, S_values[i]))
            S_values[i] = sprt(incorrect, a, r, S_values[i])
            #print(S_values[i])
        for i in range(n):
            if S_values[i] >= np.log(1/deltas[i]):
                found = True
                concept = hypotheses[i]
       # print(run)
        min_errors = np.min(num_errors)
        
        if hypotheses[-1].classify(example) != curr_class and (found_true and found_false):
            cap = 1000
            if eta != 0:
                cap = 2/eta
            new_hyp, error_num = find_noisy_consistent_halfspace(examples, min_errors, cap, classifications, dimension)
            if error_num != -1:
                hypotheses = np.append(hypotheses, new_hyp)
                n += 1
                # print(n)
                #deltas = np.append(deltas, 6 * (delta/2) / (np.pi ** 2) * (a/r) ** num_errors * ((1-a)/(1-r)) ** (run - num_errors))
                deltas = np.append(deltas, 6 * (delta) / (np.pi ** 2 * n ** 2))
                S_values = np.append(S_values, 0)
                num_errors = np.append(num_errors, error_num)
            #else:
                #print("AISJFOIEJ")
        
        #print('--------')
        
    return concept, examples
```

```python
def get_noisy_bound(epsilon, delta, eta, dimension):
    return (2 / epsilon ** 2) / (1 - 2*eta) ** 2 * np.log(2 * (2 * 10 ** 4) ** dimension * (2 * 10 ** 4) * 2 /delta)
```

```python
def make_good_interval(noisy_interval):
    params, greater, b, eta = noisy_interval.get_params()
    return halfspace(len(params), params, greater, b)
```

```python
def get_finite_halfspace(dimension):
    greater = np.random.uniform(0, 1) > 0.5
    params = np.round(np.random.uniform(-100, 100, dimension), decimals=2)
    b = np.round(np.random.uniform(-100, 100), decimals = 2)
    return halfspace(dimension, np.array(params), greater, b)
```

```python
def get_finite_noisy_halfspace(dimension, eta):
    greater = np.random.uniform(0, 1) > 0.5
    params = np.round(np.random.uniform(-100, 100, dimension), decimals=2)
    b = np.round(np.random.uniform(-100, 100), decimals = 2)
    return noisy_halfspace(dimension, np.array(params), greater, b, eta)
```

```python
def get_sample_point(dimension):
    return np.array(np.random.uniform(-10, 10, dimension))
```
